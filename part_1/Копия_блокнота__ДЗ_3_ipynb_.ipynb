## Задание 2.1 (5 баллов)

[Смещения в нуклеотидном составе](https://ru.wikipedia.org/wiki/GC-%D1%81%D0%BE%D1%81%D1%82%D0%B0%D0%B2) можно вычислять для разных пар нуклеотидов.

Используя формулы ниже, собрать информацию по смещениям на каждые 100 элементов последовательности.


$$
GC\_skew = \dfrac{G-C}{G+C}
$$
$$
AT\_skew = \dfrac{A-T}{A+T}
$$
$$
SW\_skew = \dfrac{S-W}{L}, \quad S=G+C, \quad W=A+T
$$
$$
RY\_skew = \dfrac{R-Y}{L}, \quad R=G+A, \quad Y=C+T
$$
$$
MK\_skew = \dfrac{M-K}{L}, \quad M=C+A, \quad K=G+T
$$

$A$, $T$, $G$, $C$ - количество соответсвующих нуклеотидов в последовательности длины $L$. $L = A+T+G+C$ - длина цепи.

Необходимо построить график изменения частоты на каждые 100 нуклеотидов всех указанных характеристик.

Для генерации последовательности нуклеотидов запустите следующий код. В качестве сида генерации используйте свой номер ису. Далее необходимо использовать информацию из переменной `sequence`

```python
import random
from typing import Optional

def generate_dna(length: int,
                gc_content: float = 0.5,
                seed: Optional[int] = None) -> str:
    """
    Генерирует последовательность ДНК с заданными параметрами

    Args:
        length: длина последовательности
        gc_content: содержание GC-пар (от 0.0 до 1.0)
        seed: seed для random (для воспроизводимости)

    Returns:
        Строка с последовательностью ДНК
    """
    if seed is not None:
        random.seed(seed)

    if not 0 <= gc_content <= 1:
        raise ValueError("GC content должен быть между 0 и 1")

    # Рассчитываем вероятности для каждого основания
    gc_prob = gc_content / 2  # Вероятность для G и C
    at_prob = (1 - gc_content) / 2  # Вероятность для A и T

    bases = ['a', 'g', 'c', 't']
    probabilities = [at_prob, gc_prob, gc_prob, at_prob]

    return ''.join(random.choices(bases, weights=probabilities, k=length))

# Укажите свой номер ИСУ
isu_number = 502367
sequence = generate_dna(20000, gc_content=0.6, seed=isu_number)
```
