## Задача 1. Королевский гамбит (1 балл)

Петя любит играть в шахматы. Он совершает свои первые шаги в игре и изучает правила перемещения фигур на доске. Помогите Пете ответить на вопрос: могла ли фигура совершить такое перемещение из точки $A$ в точку $B$?

В первой строке программе подается название фигуры:
- ферзь
- король
- конь
- ладья

После этого, на первой строке через пробел указываются координаты точки $A(x_1 ,y_1)$, затем на следующей координаты $B(x_2, y_2)$.

Программа должна вывести Yes, если фигура за один ход могла перейти из точки $A$ в точку $B$, и No в противном случае.

Считайте, что поля пронумерованы от 1 до 8, начиная с левого нижнего угла.

Пример ввода:
```
король
1 2
5 7
```

Пример вывода:
```
No
```

````python
figure = input().strip()

x1, y1 = map(int, input().split())
x2, y2 = map(int, input().split())

dx = abs(x2 - x1)
dy = abs(y2 - y1)

if figure == 'ферзь':
  result = (x1 == x2 or y1 == y2 or dx ==dy)
elif figure == 'король':
  result = (dx <= 1 and dy <= 1 and (dx != 0 or dy != 0))
elif figure == 'конь':
  result = (dx == 1 and dy == 2) or (dx == 2 and dy == 1)
elif figure == 'ладья':
  result = (x1 == x2 or y1 == y2)

print("Yes" if result else "No")
print("Ты вне поля" if (x1 < 1 or y1 < 1 or x2 < 1 or y2 < 1 or
                         x1 > 8 or y1 > 8 or x2 > 8 or y2 > 8) else "Ты в поле")
````

Пользователь на вход вводит значение $n$. В ответе необходимо найти значение выражения
$$
1! + 2! + \dots + n!
$$

Пользоваться математической библиотекой math в этой задаче запрещено.

Пример ввода:
```

```

Пример вывода:
```

```

````python
n = int(input().strip())
if n == 0:
     print(1)
else:
    total = 0
    fact = 1
for i in range(1, n + 1):
        fact *= i
        total += fact
print(total)
````

## Задача 6. Снежинка (1 балл)

По данному нечетному числу $n$ необходимо создать двумерный массив, образующий снежинку. После чего необходимо ее вывести в консоль. На главной и побочной диагоналях, а также в средней строке и среднем столбце стоят символы "*", в остальных ячейках - точки "."


Пример ввода:
```
9
```

Пример вывода:
```
* . . . * . . . *
. * . . * . . * .
. . * . * . * . .
. . . * * * . . .
* * * * * * * * *
. . . * * * . . .
. . * . * . * . .
. * . . * . . * .
* . . . * . . . *
```

````python
n = int(input())
snowflake = [['.' for _ in range(n)] for _ in range(n)]
for i in range(n):
    for j in range(n):
        if i == j or i + j == n - 1 or i == n // 2 or j == n // 2:
            snowflake[i][j] = '*'
for row in snowflake:
    print(' '.join(row))
````

## Задача 7. Ветер перемен (1 балл)

Дан двумерный массив и два числа: $i$ и $j$. Поменяйте в массиве столбцы с номерами $i$ и $j$ и выведите результат.

Программа получает на вход размеры массива $n$ и $m$, затем элементы массива, затем числа $i$ и $j$.

Решение оформите в виде функции `swap_columns(a, i, j)`.

Пример ввода:
```
4 6
11 12 13 14 15 16
21 22 23 24 25 26
31 32 33 34 35 36
41 42 43 44 45 46
0 5
```

Пример вывода:
```
16 12 13 14 15 11
26 22 23 24 25 21
36 32 33 34 35 31
46 42 43 44 45 41
```

````python
def swap_columns(a, i, j):
    for row in a:
        row[i], row[j] = row[j], row[i]
n, m = map(int, input().split())
a = []
for _ in range(n):
    row = list(map(int, input().split()))
    a.append(row)
i, j = map(int, input().split())
swap_columns(a, i, j)
for row in a:
    print(*row)
````

## Задча 9. Едим дома (1 балл)

Каждый житель некоторой страны любит неоторые блюда. Нужно определить какие в этом обществе общие вкусовые предпочтения, а также список всех существуюзих гастрономических изысков.

В первой строке программа получает число $n$ - число жителей города. Далее по каждому человеку сначала передается информация о количестве его любимых блюд $k_j$, после чего друг за другом указываются названия этих блюд.

В ответе укажите сперва количество блюд, которые нравятся всем жителям, если они есть, то далее в алфавитном порядке перечислите названия этих блюд по одному в строке. Затем укажите количество существующих в стране блюд и также в лексикографическом порядке перечислите их.


Пример ввода:
```
3
2
Лазанья
Пицца
3
Суши
Пицца
Пудинг
1
Пицца
```

Пример вывода:
```
1
Пицца
4
Лазанья
Пицца
Пудинг
Суши
```

````python
n = int(input())

common_dishes = None
all_dishes = set()

for _ in range(n):
    k = int(input())
    current = set()
    for _ in range(k):
        dish = input().strip()
        current.add(dish)

    all_dishes |= current

    if common_dishes is None:
        common_dishes = current
    else:
        common_dishes &= current

common_sorted = sorted(common_dishes)
all_sorted = sorted(all_dishes)

print(len(common_sorted))
for dish in common_sorted:
    print(dish)

print(len(all_sorted))
for dish in all_sorted:
    print(dish)
````

## Задача 10. Англо-латинский словарь (1 балл)

Однажды, разбирая старые книги на чердаке, школьник Вася нашёл англо-латинский словарь. Английский он к тому времени знал в совершенстве, и его мечтой было изучить латынь. Поэтому попавшийся словарь был как раз кстати.

К сожалению, для полноценного изучения языка недостаточно только одного словаря: кроме англо-латинского необходим латинско-английский. За неимением лучшего он решил сделать второй словарь из первого.

Как известно, словарь состоит из переводимых слов, к каждому из которых приводится несколько слов-переводов. Для каждого латинского слова, встречающегося где-либо в словаре, Вася предлагает найти все его переводы (то есть все английские слова, для которых наше латинское встречалось в его списке переводов), и считать их и только их переводами этого латинского слова.

Помогите Васе выполнить работу по созданию латинско-английского словаря из англо-латинского.

В первой строке содержится единственное целое число N — количество английских слов в словаре. Далее следует N описаний. Каждое описание содержится в отдельной строке, в которой записано сначала английское слово, затем отделённый пробелами дефис, затем разделённые запятыми с пробелами переводы этого английского слова на латинский. Все слова состоят только из маленьких латинских букв. Переводы отсортированы в лексикографическом порядке. Порядок следования английских слов в словаре также лексикографический.

Выведите соответствующий данному латинско-английский словарь, в точности соблюдая формат входных данных. В частности, первым должен идти перевод лексикографически минимального латинского слова, далее — второго в этом порядке и т.д. Внутри перевода английские слова должны быть также отсортированы лексикографически.


Пример ввода:
```
3
apple - malum, pomum, popula
fruit - baca, bacca, popum
punishment - malum, multa
```

Пример вывода:
```
7
baca - fruit
bacca - fruit
malum - apple, punishment
multa - punishment
pomum - apple
popula - apple
popum - fruit
```


````python
n = int(input())
latin_to_english = {}

for _ in range(n):
    line = input().strip()
    eng_word, latin_part = line.split(' - ')
    latin_words = latin_part.split(', ')
    for lat in latin_words:
        if lat not in latin_to_english:
            latin_to_english[lat] = set()
        latin_to_english[lat].add(eng_word)

latin_sorted = sorted(latin_to_english.keys())

print(len(latin_sorted))
for lat in latin_sorted:
    eng_sorted = sorted(latin_to_english[lat])
    print(f"{lat} - {', '.join(eng_sorted)}")
````
